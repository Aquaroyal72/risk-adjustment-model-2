```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = pd.read_csv("MarketExitAll_no2000.csv")

for_profit = data.where(data["Not-For-Profit:_2014"] == "No").dropna()
non_profit = data.where(data["Not-For-Profit:_2014"] == "Yes").dropna()
```

```python
plt.hist(for_profit["IndRTPMM"] - for_profit["IndRTPMM_2014"], 50, facecolor='g')


plt.xlabel('Transfer Difference')
plt.title('Individual Group Risk Transfer Difference between 2016 and 2014 BY in For-Profit Companies')
plt.grid(True)
plt.show()
```

```python
plt.hist(non_profit["IndRTPMM"] - non_profit["IndRTPMM_2014"], 50, facecolor='g')


plt.xlabel('Transfer Difference')
plt.title('Individual Group Risk Transfer Difference between 2016 and 2014 BY in Non-Profit Companies')
plt.grid(True)
plt.show()
```

```python
for_profit.to_csv("forprofit.csv")
```

```python
non_profit.to_csv("nonprofit.csv")
```
