```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats.stats import pearsonr
```

```python
data = pd.read_csv("Output/ratiosenhanced.csv")[["state", "individualratio", 
                        "smallgroupratio", "PLRSindividual2015", "PLRSsmallgroup2015",
                             "PLRSindividual2016", "PLRSsmallgroup2016"]]
```

```python
data.head()
```

```python
fig, ax1 = plt.subplots(figsize=(7, 7))
ax1.set_title('Individual PLRS 2015 vs. Individual Transfers:MM Ratio')

ax1.plot(data["PLRSindividual2015"], 
         data["individualratio"], '+', color='r')
ax1.set_ylabel('individual Transfers:MM ratio')
ax1.set_xlabel('individual PLRS 2015')

#ax1.set_xlim([-10000,5000])
#ax1.set_ylim([-500,500])
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(7, 7))
ax1.set_title('Individual PLRS 2016 vs. Individual Transfers:MM Ratio')

ax1.plot(data["PLRSindividual2016"], 
         data["individualratio"], '+', color='r')
ax1.set_ylabel('individual Transfers:MM ratio')
ax1.set_xlabel('individual PLRS 2016')

#ax1.set_xlim([-10000,5000])
#ax1.set_ylim([-500,500])
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(7, 7))
ax1.set_title('Difference in Individual PLRS 2016-2015 vs. Individual Transfers:MM Ratio')

ax1.plot(data["PLRSindividual2016"] - data["PLRSindividual2015"], 
         data["individualratio"], '+', color='r')
ax1.set_ylabel('individual Transfers:MM ratio')
ax1.set_xlabel('difference in individual PLRS 2016-2015')

#ax1.set_xlim([-10000,5000])
#ax1.set_ylim([-500,500])
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(7, 7))
ax1.set_title('Small Group PLRS 2015 vs. Small Group Transfers:MM Ratio')

ax1.plot(data["PLRSsmallgroup2015"], 
         data["smallgroupratio"], '+', color='b')
ax1.set_ylabel('small group Transfers:MM ratio')
ax1.set_xlabel('small group PLRS 2015')

#ax1.set_xlim([-10000,5000])
#ax1.set_ylim([-500,500])
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(7, 7))
ax1.set_title('Small Group PLRS 2016 vs. Small Group Transfers:MM Ratio')

ax1.plot(data["PLRSsmallgroup2016"], 
         data["smallgroupratio"], '+', color='b')
ax1.set_ylabel('small group Transfers:MM ratio')
ax1.set_xlabel('small group PLRS 2016')

#ax1.set_xlim([-10000,5000])
#ax1.set_ylim([-500,500])
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(7, 7))
ax1.set_title('Difference in Small Group PLRS 2016 - 2015 vs. Small Group Transfers:MM Ratio')

ax1.plot(data["PLRSsmallgroup2016"] - data["PLRSsmallgroup2015"], 
         data["smallgroupratio"], '+', color='b')
ax1.set_ylabel('small group Transfers:MM ratio')
ax1.set_xlabel('difference in small group PLRS 2016 - 2015')

#ax1.set_xlim([-10000,5000])
#ax1.set_ylim([-500,500])
plt.show()
```

```python
print(pearsonr(pd.to_numeric(data["PLRSindividual2015"]), pd.to_numeric(data["individualratio"])))
print(pearsonr(pd.to_numeric(data["PLRSindividual2016"]), pd.to_numeric(data["individualratio"])))
print(pearsonr(pd.to_numeric(data["PLRSindividual2016"] - data["PLRSindividual2015"]), pd.to_numeric(data["individualratio"])))
print(pearsonr(pd.to_numeric(data["PLRSsmallgroup2016"]), pd.to_numeric(data["smallgroupratio"])))
print(pearsonr(pd.to_numeric(data["PLRSsmallgroup2015"]), pd.to_numeric(data["smallgroupratio"])))
print(pearsonr(pd.to_numeric(data["PLRSsmallgroup2016"] - data["PLRSsmallgroup2015"]), pd.to_numeric(data["smallgroupratio"])))
```
