Absolute transfers over premiums per state:


```python
import pandas as pd
```

```python
transfers = pd.read_csv("absolutetransfersperstate2.csv", low_memory=False)
plrs = pd.read_csv("2017data.csv", low_memory=False)
```

```python

transfers = transfers.apply(pd.to_numeric, errors='ignore')
```

```python
transfers.head()
```

```python
plrs.head()
```

```python
states = plrs["State"]
cols = ['State', 'Absolute Risk Adjustment Individual Market', 'Absolute Risk Adjustment Small Group Market']
payment_sum = pd.DataFrame(columns=cols)
payment_sum.loc[0] = [1.0, 'hotel', 'true']
print(payment_sum)
```

```python
i = 0
for state in states:
    temp = transfers.where(transfers["STATE"] == state).dropna()
    a = sum(temp["ABSOLUTE RISK ADJUSTMENT INDIVIDUAL MARKET"].convert_objects(convert_numeric=True).dropna())
    b = sum(temp["ABSOLUTE RISK ADJUSTMENT SMALL GROUP MARKET"].convert_objects(convert_numeric=True).dropna())
    new_row = [state, a, b]
    payment_sum.loc[i] = new_row
    print(state + " " + str(a) + " " + str(b))
    i = i + 1
```

```python
payment_sum.head()
```

```python
total = pd.merge(payment_sum, plrs, on='State')
```

```python
total.head()
```

```python
print("state,individual_ra_premiums_ratio,smallgroup_ra_premiums_ratio")
for state in states:
    thisstate = total.where(total["State"] == state).dropna()
    individual = pd.to_numeric(thisstate["Absolute Risk Adjustment Individual Market"].item()
                               / thisstate["STATE BILLABLE MEMBER MONTHS Individual"].item()
                               / thisstate["STATE AVERAGE MONTHLY PREMIUMS Individual"].item())
    smallgroup = pd.to_numeric(thisstate["Absolute Risk Adjustment Small Group Market"].item()
                               / thisstate["STATE BILLABLE MEMBER MONTHS Small Group"].item()
                               / thisstate[" STATE AVERAGE MONTHLY PREMIUMS Small Group"].item())
    print(state + "," + str(individual) + "," + str(smallgroup))
```
